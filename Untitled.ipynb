```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score,calinski_harabasz_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.cluster import Birch
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load the dataset
path = "kaggle_Interests_group/kaggle_Interests_group.csv"
data = pd.read_csv(path)
data.head()
```

```python

# Replace NaN with 0, assuming NaN means no interest
data_filled = data.fillna(0)

# Encode the 'group' column numerically if necessary for visualization or further analysis
group_encoded = pd.get_dummies(data_filled['group'], drop_first=True)

# Prepare data for scaling
# Dropping 'group' and 'grand_tot_interests' as they are not necessary for clustering
features = data_filled.drop(columns=['group', 'grand_tot_interests'])

# Apply StandardScaler for normalization
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Show the scaled features
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_features_df.head()
```

```python
# Splitting the dataset into training, evaluation, and testing sets
# First, split into training and temporary set (combining evaluation and testing)
train_data, temp_data = train_test_split(scaled_features_df, test_size=0.3, random_state=42)

# Now, split the temporary set into evaluation and testing sets
eval_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# Display the sizes of each dataset to verify
(train_data.shape, eval_data.shape, test_data.shape)
```

```python

# Initialize K-means clustering model with 4 clusters
kmeans_model = KMeans(n_clusters=4, random_state=42)

# Fit the K-means model on the training data
kmeans_model.fit(train_data)

# Display the cluster centers
kmeans_model.cluster_centers_
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features_df)

# Predict the cluster labels for the dataset
cluster_labels = kmeans_model.predict(scaled_features_df)

# Plotting the results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.title('Visualization of K-Means Clustering with 2D PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

```python
silhouette_avg = silhouette_score(scaled_features_df, cluster_labels)
davies_bouldin = davies_bouldin_score(scaled_features_df, cluster_labels)
calinski_harabasz = calinski_harabasz_score(scaled_features_df, cluster_labels)

(silhouette_avg, davies_bouldin, calinski_harabasz)
```

```python
# using elbow method to improve the kmeans model
# Calculate distortion for a range of number of cluster
inertia = []
k_range = range(1, 11)  # Testing from 1 to 10 clusters

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features_df)
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(10, 8))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```
